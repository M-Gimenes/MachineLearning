```python
import pandas as pd
table = pd.read_csv("../0-Data/2015-building-energy-benchmarking.csv")
pd.set_option("display.max_columns", len(table.columns))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (8,8)).set_facecolor("gray")
sns.heatmap(table.corr(method = "pearson", numeric_only=True))
plt.show()
```

```python
import pandas as pd
table = pd.read_csv("../0-Data/traffic-collision-data-from-2010-to-present.csv")
pd.set_option("display.max_columns", len(table.columns))
table.head()
```

```python
area_encode = pd.get_dummies(table["Area Name"], dtype=float)
area_encode.head()
```

```python
concatenated = pd.concat([table, area_encode], axis = 1)
pd.set_option("display.max_columns", len(concatenated.columns))
concatenated.drop("Area Name", axis = 1, inplace = True)
concatenated.head()
```

```python
import matplotlib.pyplot as plt
concatenated.boxplot(column="Census Tracts")
plt.show()
```

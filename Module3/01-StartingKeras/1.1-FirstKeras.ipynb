```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
x = pd.DataFrame(iris.data, columns=[iris.feature_names])
y = pd.Series(iris.target)

x.head()
```

```python
y.value_counts()
```

```python
import keras

y_converted = keras.utils.to_categorical(y)
y_converted
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y_converted, test_size=0.3)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
model = Sequential()
model.add(Dense(10, input_dim=4, kernel_initializer='normal', activation='relu'))
model.add(Dense(3, kernel_initializer='normal', activation='softmax'))
```

```python
optimizer = SGD()
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['acc'])
```

```python
model.fit(x_train, y_train, epochs=500, batch_size=60, validation_data=(x_test, y_test), verbose=1)
```

```python
predicts = model.predict(x_test)

import numpy as np
np.set_printoptions(formatter={'float': lambda x: f"{0:0.2f}"})
print(predicts)
```
